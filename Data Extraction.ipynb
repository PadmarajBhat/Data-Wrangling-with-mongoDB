##### Exercise: Identify the different types 

```python
filename = 'cities.csv'
FIELDS = ["name", "timeZone_label", "utcOffset", "homepage", "governmentType_label",
          "isPartOf_label", "areaCode", "populationTotal", "elevation",
          "maximumElevation", "minimumElevation", "populationDensity",
          "wgs84_pos#lat", "wgs84_pos#long", "areaLand", "areaMetro", "areaUrban"]
field_types = {}
index_field = {}
with open(filename,'r', encoding='utf-8') as f:
    #get the field number pairing
    all_fields = [x.strip() for x in f.readline().split(',')]
    print(all_fields)
    for n in FIELDS:
        index_field[n] = all_fields.index('"'+n+'"')
    
    f.readline()
    f.readline()
    f.readline()
    
    for line in f:
        token = line.split(',')
        for f in FIELDS:
            if token[index_field[f]] == "NULL":
                field_types[f].add(type(None))
            else:
                if f in field_types:
                    field_types[f].add(type(None))
                else:
                    field_types[f]=set()
                    field_types[f].add(type(token[index_field[f]]))
                
            
field_types            
```
